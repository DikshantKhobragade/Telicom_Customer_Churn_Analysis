## Converted 0 and 1 values of senior citizen to yes/no

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"
df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})

plt.pie(gb['Churn'],labels = gb.index,autopct= "%1.2f%%")
plt.title("percentage Of Churned Customers",fontsize = 10)
plt.show()
```

```python
plt.figure(figsize =(6,4))
sns.countplot(x = "gender", data=df, hue = "Churn" )
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize =(6,4))
ax = sns.countplot(x = "SeniorCitizen", data=df )
ax.bar_label(ax.containers[0])
plt.title("Churn of customers by SeniorCitizen")
plt.show()
```

```python

grouped_data = df.groupby(['SeniorCitizen', 'Churn']).size().reset_index(name='Count')
pivot_data = grouped_data.pivot(index='SeniorCitizen', columns='Churn', values='Count')
pivot_data = pivot_data.fillna(0)

# Convert to percentages
pivot_percentage = pivot_data.div(pivot_data.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
fig, ax = plt.subplots(figsize=(6, 4))
pivot_percentage.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e'])

# Add percentage labels
for i in range(pivot_percentage.shape[0]):  # Loop over each bar group
    cumulative_percentage = 0
    for j in range(pivot_percentage.shape[1]):  # Loop over segments in a group
        percentage = pivot_percentage.iloc[i, j]
        cumulative_percentage += percentage
        if percentage > 0:  # Avoid labels for empty segments
            ax.text(i, cumulative_percentage - percentage / 2, f'{percentage:.1f}%', ha='center', va='center', fontsize=10, color='white')

# Customize chart
ax.set_title("Churn by SeniorCitizen (Percentage)", fontsize=14)
ax.set_xlabel("SeniorCitizen", fontsize=12)
ax.set_ylabel("Percentage", fontsize=12)
ax.legend(title="Churn", loc='upper right')
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

```python

# Define the columns you want to plot
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Number of subplots (3 rows and 3 columns for 9 columns)
n_cols = 3
n_rows = -(-len(columns) // n_cols)  # Calculate rows with ceiling division

# Create the subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 10))
axes = axes.flatten()  # Flatten the axes array for easy iteration

# Plot each column
for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i],hue = df["Churn"])
    axes[i].set_title(f'Countplot of {col}', fontsize=12)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Remove unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```

# The majority of customers who do not churn tend to have services like PhoneService, InternetService (particularly DSL), and OnlineSecurity enabled. For services like OnlineBackup, TechSupport, and StreamingTV, churn rates are noticeably higher when these services are not used or are unavailable.

## customer is likely to churn when he is using electronic check as a payment method.

### **Insights**
1. **Customer Tenure:** Customers with shorter tenure are more likely to churn. Retention programs targeted at newer customers can be effective in reducing churn.
2. **Monthly Charges:** High monthly charges are a key driver of churn. Consider offering discounts or tailored pricing plans for customers in higher charge brackets.
3. **Contract Type:** Customers on month-to-month contracts are more likely to churn compared to those on long-term contracts. Offering incentives for customers to switch to annual or longer-term contracts could improve retention.
4. **Payment Method:** Electronic check users have a higher churn rate. Encouraging the adoption of more stable payment methods, such as credit cards or automatic payments, may reduce churn.
5. **Demographics:** Certain demographic factors (e.g., senior citizens) may influence churn. Tailored services for specific customer segments could increase satisfaction.


### **Recommendations**

1. **Onboarding Programs:**
   - Implement a robust onboarding process to engage new customers and address their needs early on.
   - Regular check-ins during the first few months can improve the experience for new users.

2. **Flexible Pricing Options:**
   - Offer flexible or customized pricing plans to high-risk customers (e.g., those with high monthly charges).
   - Introduce loyalty rewards or discounts for long-term customers.

3. **Promotions for Contract Upgrades:**
   - Provide incentives (e.g., discounts or added benefits) for customers to move from month-to-month contracts to long-term agreements.

4. **Payment Method Adoption:**
   - Promote secure and automatic payment methods to reduce churn among electronic check users.
   - Offer a one-time discount or benefits for customers who switch to more stable payment methods.
